# Descriptive Analysis

### Descriptive Statistics
for each column:
    - what's the min, the max, the mean, the std, the missingness (%), unique values (to detect quasi categorical values)

```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_parquet("hf://datasets/MoSBAIHI/weld-quality-dataset/data/train-00000-of-00001.parquet")
df = df.drop(columns=["Weld ID"])
df = df.replace("N", None)
df.head(n=3)
```

```python
def column_stats(df, col):
    series = df[col]
    stats = {}
    stats['max'] = pd.to_numeric(series, errors='coerce').max()
    stats['min'] = pd.to_numeric(series, errors='coerce').min()
    stats['mean'] = pd.to_numeric(series, errors='coerce').mean()
    stats['std'] = pd.to_numeric(series, errors='coerce').std()
    stats['missingness'] = series.isna().mean() * 100
    stats['n_unique'] = series.nunique(dropna=True)
    return stats
```

```python
stats = {}
for col in df.columns:
    stats[col] = column_stats(df, col)
stats_df = pd.DataFrame(stats).T
stats_df.head(n=10)
```
